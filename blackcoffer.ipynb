```python
def createnew(url):
    import requests
    bc = requests.get(url)
    from bs4 import BeautifulSoup
    from nltk.tokenize import sent_tokenize, word_tokenize
    bs = BeautifulSoup(bc.content, 'html.parser')
    words = bs.findAll(attrs={'class': 'td-post-content tagdiv-type'})

    if len(words) > 0:
        text = words[0].text.replace("\n", "")
        import re
        text = re.sub(r'\[d+\]', "", text)
        textsent = sent_tokenize(text)
        return ' '.join(textsent), text
    else:
        return "", ""
```

```python
def count_syllables(word):
    vowels = "aeiouy"
    count = 0
    last_char_was_vowel = False

    for char in word:
        if char.lower() in vowels:
            if not last_char_was_vowel:
                count += 1
            last_char_was_vowel = True
        else:
            last_char_was_vowel = False

    if word.endswith("e"):
        count -= 1

    if count == 0:
        count = 1

    return count

def word_complexity(word):
    length = len(word)
    syllables = count_syllables(word)
    return length + syllables

```

def percent_complexity(text):
    words = text.split()
    complex_word_count = sum(1 for word in words if count_syllables(word) > 2)
    return (complex_word_count / len(words)) * 100 if len(words) > 0 else 0

```python
def percentcomplex(text):
    text_str = ' '.join(text)
    complex_words = complexity(text_str)
    if complex_words is None:
        return 0
    return complex_words / len(text)
```

def percentcomplex1(text):
    complex_words = complexity(text)
    if complex_words is None:
        return 0
    return complex_words / len(text.split())

```python
import nltk
from nltk import word_tokenize
def personpro(sentence):
    pronouns = ["I", "me", "my", "mine", "myself",
                "you", "your", "yours", "yourself",
                "he", "him", "his", "himself",
                "she", "her", "hers", "herself",
                "it", "its", "itself",
                "we", "us", "our", "ours", "ourselves",
                "they", "them", "their", "theirs", "themselves"]

    tokens = word_tokenize(sentence)
    count = sum(1 for token in tokens if token.lower() in pronouns)
    return count
```

```python
def negative(token):
    n=0
    with open('negative-words.txt','r') as neg:
        nw=neg.readlines()
        nwo=[i.rstrip('\n') for i in nw]
    for w in token:
        if w in nwo:
            n=n+1
    return n
```

```python
def positive(token):
    n=0
    with open('positive-words.txt','r') as pos:
        pw=pos.readlines()
        pwo=[i.rstrip('\n') for i in pw]
    for w in token:
        if w in pwo:
            n=n+1
    return n
```

```python
def polarity(tokenword):
    return ((positive(tokenword)-negative(tokenword))/(positive(tokenword)+negative(tokenword))+0.0000001)
```

```python
def complexity(tokenword):
    cc=0
    for w in tokenword:
        vc=0
        if w.endswith(('es','ed')):
            continue
        else:
            for wc in w:
                if(wc=='a' or wc=='e' or wc=='i' or wc=='o' or wc=='u'):
                    vc=vc+1
            if(vc>2):
                cc=cc+1
        return cc
```

```python
def asl(tokenword,tokensent):
    if len(tokensent)>0:
        return len(tokenword)/len(tokensent)
    else:
        return 0
```

```python
def fogindex(tokenword,tokensent):
    return 0.4*asl(tokenword,tokensent)+percentcomplex(tokenword)
```

```python
def avgwdl(tokenword):
    l=[]
    s=0
    for wc in tokenword:
        l.append(len(wc))
        s+=len(wc)
    if len(tokenword)>0:
        return s/len(tokenword)
    else:
        return 0
```
